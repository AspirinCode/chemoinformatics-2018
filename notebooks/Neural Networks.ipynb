```python
import numpy as np
from random import shuffle
import matplotlib.pyplot as plt

%matplotlib inline
```

### 1. Generate dummy data

```python
def linear_dummy_data(n=500, w=10, b=30, noise=0.1, seed=0):
    np.random.seed(seed)
    x = np.random.randn(n)
    y = w*x + b + noise*np.random.randn(n)
    
    return x, y
```

### 2. Simple Linear Regression Model

```python
# Generate data
conf = {
    'n':1000,
    'w': 10,
    'b': 30,
    'noise': 2.0,
}

x, y = linear_dummy_data(**conf)

# Plot
figsize = (16, 9)
fig, ax = plt.subplots(figsize=figsize)
ax.scatter(x, y, marker="o")
ax.plot(x, conf['w']*x + conf['b'], color='r')
ax.axes.set_xlabel('x')
ax.axes.set_ylabel('y')
```

```python
# Gradient descent
w_pred = np.random.rand(1)
b_pred = np.random.rand(1)

# Optimization paramters
num_epochs = 500
learning_rate = 0.00005

losses = list()
for i in range(num_epochs):
    shuffle([x, y])
    
    for (x_point, y_point) in zip(x, y):
        y_pred = w_pred*x_point + b_pred
        loss = np.power(y_point - y_pred, 2)
        
        dw = -2 * (y_point - y_pred) * x_point
        db = -2 * (y_point - y_pred)
        
        w_pred -= learning_rate * dw
        b_pred -= learning_rate * db
        
    if i % 50 == 0:        
        print(f"Epoch: {i}")
        print(f"Loss: {loss}, W: {w_pred}, b: {b_pred}\n")
        losses.append(loss)
```

```python
figsize = (16, 10)
fig, ax = plt.subplots(figsize=figsize)
ax.plot(losses, marker="o")
ax.axes.set_xlabel("Number of epochs")
ax.axes.set_ylabel("Loss")
#ax.axes.set_xticks(list(range(num_epochs)));
```
